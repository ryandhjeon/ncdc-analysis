```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./outputs/out_task02")
df.columns = ['year', 'min_temp', 'max_temp', 'avg_temp', 'median_temp']

col_names = df.columns

for col in col_names:
    df[col] = df[col].map(lambda x: str(x).lstrip('null\t[').rstrip(']')).astype(float)

df = df.astype({'year': int})    
df.head()
```

```python
df = df.melt('year', var_name='cols',  value_name='temps')
df.describe()
```

```python
plt.figure(figsize = (15,8))

sns.lineplot(x="year", y="temps", hue='cols', data=df, palette = "hls")

plt.title("Temperatures for year 1920 ~ 1940", fontsize = 20)
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Temperature", fontsize = 15)

plt.show()
```
